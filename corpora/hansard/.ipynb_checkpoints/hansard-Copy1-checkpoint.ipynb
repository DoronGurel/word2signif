# hansard scraper

the script below is responsable of scraping the hansard archive and organizing the text by politica party, in an ordarly fashion suitable for gensim word2vec function to train on

### imports

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random
import gensim
import nltk
import re
#nltk.download('punkt')
#nltk.download('stopwords')
import pickle
import os
from tqdm import tqdm
import shutil
from gensim.models import translation_matrix
from gensim.models import KeyedVectors
import seaborn as sns
import matplotlib.pyplot as plt
```

### functions for scraping and data extraction

```python
all_mp_url = "https://hansard.parliament.uk/search/Members?currentFormerFilter=1&house=Commons&startDate=2013-01-01&endDate=2019-05-24&partial=False"

def results_2_links(soup):
    results = soup.find('div', class_="results-list row")
    results_inner = results.find_all('div', class_="col-sm-6 result-outer")
    debate_links = []
    for i in range(len(results_inner)):
        debate_links.append(results_inner[i].find('a')['href'])
    return(debate_links)

def debate_2_data(debate):
    debate_content_list = debate.find_all("div", class_="content-item")
    data = []
    for block in debate_content_list:
        if block.find("h2", class_="memberLink"):
            if block.find("h2").find("a"):
                mp_name = block.find("h2").find("a").text
                mp_id = block.find("h2").find("a")['href'].split("=")[-1]
                content_block = block.find_all("p")
                content = ""
                for p in content_block:
                    if p.contents:
                        content += p.text

                data.append((mp_name, mp_id, content))
    
    data_df = pd.DataFrame(data)
    return data_df

def collect_debates(start_date, end_date):
    all_debates_url = 'https://hansard.parliament.uk/search/Debates?endDate={end_date}&house=Commons&partial=True&startDate={start_date}&page={page}'
    hansard_link = 'https://hansard.parliament.uk'
    all_data = []
    for i in range(1,10000):

        #query hansard
        r = requests.get(all_debates_url.format(start_date=start_date, end_date=end_date,page=str(i)))
        soup = BeautifulSoup(r.content, "lxml")

        # stop if no results
        if soup.find("div", class_="results-heading no-results" ):
            break    

        #get results
        debate_links = results_2_links(soup)
        for link in debate_links:
            r = requests.get(hansard_link+link)
            debate = BeautifulSoup(r.content, "lxml")
            data = debate_2_data(debate)
            data['debate'] = link.split("/")[-1]
            all_data.append(data)
            # rest
            time.sleep(0.19)

        print("collected page #{}".format(str(i)))

    df_all_data = pd.concat(all_data)
    return df_all_data

def models2scs(model_1, model_2):
    word_vectors_1 = model_1.wv
    word_vectors_2 = model_2.wv

    vocab_1 = [key for key in word_vectors_1.vocab]
    vocab_2 = [key for key in word_vectors_2.vocab]

    vocab = list(set(vocab_1).intersection(set(vocab_2)))
    word_pair = list(zip(vocab,vocab))

    transmat = translation_matrix.TranslationMatrix(word_vectors_1, word_vectors_2, word_pair)
    transmat.train(word_pair)

    space_1 = gensim.models.translation_matrix.Space.build(word_vectors_1)
    space_1_translated = transmat.apply_transmat(space_1)

    self_similarities = []
    words_knockoff = []
    for i in range(len(vocab_1)):
        if vocab_1[i] in vocab_2:
            words_knockoff.append(vocab_1[i])
            self_similarities.append((1 - word_vectors_2.distances(space_1_translated.mat[i,:], other_words=[vocab_1[i]]))[0])

    return self_similarities, vocab

def model2wc(model):
    words = []
    counts = []
    for item in model.wv.vocab:
        words.append(item)
        counts.append(model.wv.vocab[item].count)
    w2c = {
        'word': words,
        'word_count': counts
    }
    df_w2c = pd.DataFrame.from_dict(w2c)
    return df_w2c

def knockoff_pct(df_all, min_count=100, num_examples=100):
    num_examples = min(df_all.shape[0], num_examples)
    
    df_all = df_all[(df_all['word_count_labour'] > min_count) & (df_all['word_count_conservative'] > min_count)].sort_values(by=['SCS'])
    vals = []
    nums = range(0,num_examples, 1)
    for i in nums:
        vals.append(sum(df_all['source'][0:i] == 'knockoff')/(i+1))
    
    return pd.DataFrame({'knockoof_pct':vals, 'top_n':nums})
    
```

# learn WE models

```python
class MySentences(object):
    def __init__(self, dirname_1, dirname_2):
        if dirname_1 != dirname_2:
            self.dirname_1 = dirname_1
            self.file_list_1 = [self.dirname_1 +'/'+fname for fname in os.listdir(dirname_1) if not (fname.endswith(".ipynb") or fname.endswith("ipynb_checkpoints"))] 
            #print(len(self.file_list_1))
            self.dirname_2 = dirname_2
            self.file_list_2 = [self.dirname_2 +'/'+fname for fname in os.listdir(dirname_2) if not (fname.endswith(".ipynb") or fname.endswith("ipynb_checkpoints"))] 
            #print(len(self.file_list_2))
            self.file_list = self.file_list_1 + self.file_list_2
            random.shuffle(self.file_list)
            self.file_list = self.file_list[0:len(self.file_list_1)]
        
        else:
            self.dirname_1 = dirname_1
            self.file_list = [self.dirname_1 +'/'+fname for fname in os.listdir(dirname_1) if not (fname.endswith(".ipynb") or fname.endswith("ipynb_checkpoints"))] 
            random.shuffle(self.file_list)

    def __iter__(self):
        for file in self.file_list:
            if not (file.endswith(".ipynb") or file.endswith("ipynb_checkpoints")):
                with open(file, 'rb') as pickle_file:
                    content = pickle.load(pickle_file)
                if len(content) >0:
                    yield content[0]
```
